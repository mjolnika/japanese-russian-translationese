# train classifiers

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import random
import numpy as np

from tqdm.auto import tqdm
```

```python
!gdown 1HxdMJ3ks_PwSc9tarUr6yo3-H6pje7K3
df = pd.read_csv('FINALDF.tsv', sep='\t')
df
```

```python
X = df.drop(['afile', 	'alang',	'akorp',	'astatus'], axis=1)
def s(n):
    if n == 'jp':
        return 1
    else:
        return 0

y = df['astatus'].apply(lambda x: s(x))
```

```python
seed = 200
random.seed(seed)
np.random.seed(seed)
rdstate = 140

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
classifiers = [LogisticRegression(random_state=rdstate),
               MLPClassifier(max_iter=300, random_state=rdstate),
               GaussianNB(),
               KNeighborsClassifier(), SVC(random_state=rdstate),
               DecisionTreeClassifier(random_state=rdstate)]
names = []
f1s = []
accuracies = []
for clf in tqdm(classifiers):
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    f1 = f1_score(y_test, preds, average='weighted')
    accuracy = f1_score(y_test, preds, average='micro')

    names.append(str(clf))
    f1s.append(f1)
    accuracies.append(accuracy)
```

```python
# test 1

clf_df = pd.DataFrame({'classifier': names,
                        'f1': f1s,
                        'accuracy': accuracies})
clf_df.style.highlight_max(['f1', 'accuracy'])
```

```python
# test 2
clf_df = pd.DataFrame({'classifier': names,
                        'f1': f1s,
                        'accuracy': accuracies})
clf_df.style.highlight_max(['f1', 'accuracy'])
```
