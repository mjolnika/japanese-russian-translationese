```python
japdict = defaultdict(dict)
rudict = defaultdict(dict)
folder = 'WHOLEEVERYTHING'
for file in os.listdir(folder):
    if file.replace('.txt', '') in rus:
        path = os.path.join(folder, file)
        with open(path, encoding='utf-8') as f:
            text = f.read()
            
        
        parsed = parse(text)
        
        for par in parsed:
            for p in par:
                lemma = p['lemma'].lower()
                if not rudict[lemma]:
                    rudict[lemma] = defaultdict(dict)
                    rudict[lemma]['intext'] = []
                    rudict[lemma]['count'] = 0
                    
                rudict[lemma]['count'] += 1
                rudict[lemma]['intext'].append(file)
                
    elif file.replace('.txt', '') in jap:
        
        path = os.path.join(folder, file)
        with open(path, encoding='utf-8') as f:
            text = f.read()
            
        
        parsed = parse(text)
        
        for par in parsed:
            for p in par:
                lemma = p['lemma'].lower()
                if not japdict[lemma]:
                    japdict[lemma] = defaultdict(dict)
                    japdict[lemma]['intext'] = []
                    japdict[lemma]['count'] = 0
                    
                japdict[lemma]['count'] += 1
                japdict[lemma]['intext'].append(file)
        
        
```

```python
smalljapdict = defaultdict(dict)

japdict = dict(japdict)
rudict = dict(rudict)

for k, v in japdict.items():
    
    if k in rudict.keys():
        if len(set(v['intext'])) + len(set(rudict[k]['intext']))>32:
            
            smalljapdict[k]['intext'] = list(set(v['intext']).union(set(rudict[k]['intext'])))
            
            
            smalljapdict[k]['ru'] = rudict[k]['intext']
            smalljapdict[k]['jp'] = japdict[k]['intext']
            
            
            smalljapdict[k]['jpintext'] = set(v['intext'])
            smalljapdict[k]['ruintext'] = set(rudict[k]['intext'])
            
            
            smalljapdict[k]['jpcount'] = v['count']
            smalljapdict[k]['rucount'] = rudict[k]['count']
            smalljapdict[k]['count'] =  smalljapdict[k]['jpcount'] + smalljapdict[k]['rucount']
            smalljapdict[k]['percentage'] =  smalljapdict[k]['jpcount'] /  smalljapdict[k]['count']
```

```python
def findincorpus(searchlem, foundin):
    
    folder = 'WHOLEEVERYTHING'
    
    russ = []
    japp = []
    for file in tqdm(foundin):
        if file.replace('.txt', '') in rus:
            path = os.path.join(folder, file)
            with open(path, encoding='utf-8') as f:
                text = f.read()
                
            
            parsed = parse(text)
            
            for par in parsed:
                for p in par:
                    if p['lemma'] == searchlem:
                        russ.append([file, par])
                    
        elif file.replace('.txt', '') in jap:
            
            path = os.path.join(folder, file)
            with open(path, encoding='utf-8') as f:
                text = f.read()
                
            
            parsed = parse(text)
            
            for par in parsed:
                for p in par:
                    if p['lemma'] == searchlem:
                        japp.append([file, par])
                   
    
    return russ, japp
```

```python
ad = 'буквально, возможно, вот-вот, наверное, невольно, поспешно, постепенно'
adv = ad.split(', ')

verb = 'виднеться, возникать, испытать, испытывать, исходить, напоминать, окружать, охватить, ощутить, ощущать'
verbs = verb.split(', ')

noun = ['ощущение', 'парень', 'уверенность', 'тревога']
```

```python
def display(tokenlist):
    
    striiing = ' '.join([x['form'] for x in tokenlist])
    return re.sub(' ([\.,!\?])', r'\1', striiing)
```

```python
translationesedict = defaultdict(dict)

for a in adv:
    
    print('\n')
    print(a)
    
    if not translationesedict[a]:
        translationesedict[a]['rus'] = []
        translationesedict[a]['jap'] = []
    print('-'*100)
    total = len(smalljapdict[a]['ru']) + len(smalljapdict[a]['jp'])
    percent = smalljapdict[a]['percentage']
    j_ = len(smalljapdict[a]['jp'])
    r_ = len(smalljapdict[a]['ru'])
    
    print(f'TOTAL: {total}\njp {j_}\tpercentage {round(percent, 7)}')
    rs, jn = findincorpus(a, smalljapdict[a]['intext'])
    translationesedict[a]['rus'].append(rs)
    translationesedict[a]['jap'].append(jn)

    print('- -'* 5)
    print('EXAMPLES')
    print('------RUS------')
    
    
    
    for num, r in enumerate(rs[:20]):
        print(f'{num})', display(r))
        
    print('------JP------')
    for num, r in enumerate(jn[:100]):
         print(f'{num})', display(r))
```
